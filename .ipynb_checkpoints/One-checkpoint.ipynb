```python
#import Proxy
import requests
import json
import threading
import time
import random
import re
from urllib.parse import urlencode
from random import choice



class Checker():
    def __init__(self):
        self.targets = []
        # deal with config.json
        with open("config.json") as file:
            config = json.load(file)

        with open("usernames_list.txt", "r") as file:
            self.targets = file.read().splitlines()

        self.email = config['email']
        self.username = input("Enter Username : ")
        self.password = input("Enter Password : ")
        self.list_of_ua = config.get("list_of_ua")
        # targets = config['targetUsernames']

        self.user_agent = random.sample(self.list_of_ua, 1)[0]

    def login(self):
        self.s = requests.Session()
        self.s.headers.update({
            "Accept": "*/*",
            "Accept-Language": "en-US,en;q=0.5",
            "Accept-Encoding": "gzip, deflate, br",
            "Connection": "keep-alive",
            "Host": "www.instagram.com",
            "Origin": "https://www.instagram.com",
            "Referer": "https://www.instagram.com/",
            "User-Agent": self.user_agent,
            "X-Instagram-AJAX": "1",
            "Content-Type": "application/x-www-form-urlencoded",
            "X-Requested-With": "XMLHttpRequest",
        })

        login_post = {
            "username": self.username,
            "password": self.password,
        }
        self.r = self.s.get('https://www.instagram.com/')
        self.csrf_token = re.search('(?<="csrf_token":")\w+', self.r.text).group(0)
        self.s.headers.update({"X-CSRFToken": self.csrf_token})
        time.sleep(0.1)#5 * random.random())

        url_login = "https://www.instagram.com/accounts/login/ajax/"
        self.login = self.s.post(url_login, data=login_post, allow_redirects=True)

        if self.login.status_code not in (200, 400):
            print("[{0}] Login failed!".format(self.username))
            print("Response Status: {login.status_code}")
        else:
            loginResponse = self.login.json()
            self.csrftoken = self.login.cookies["csrftoken"]
            self.s.headers.update({"X-CSRFToken": self.csrftoken})

            if loginResponse.get("errors"):
                print("Something is wrong with Instagram! Please try again later...")
                print(loginResponse["errors"]["error"])
            elif loginResponse.get("message") == "checkpoint_required":
                print('[{0}] checkpoint required!'.format(self.username))
            elif loginResponse.get("authenticated") is False:
                print("Login error! Check your login data!")
                return
            else:
                self.rollout_hash = re.search('(?<="rollout_hash":")\w+', self.r.text).group(0)
                self.s.headers.update({"X-Instagram-AJAX": self.rollout_hash})
                self.successfulLogin = True
                # ig_vw=1536; ig_pr=1.25; ig_vh=772;  ig_or=landscape-primary;
                self.s.cookies["csrftoken"] = self.csrftoken
                self.s.cookies["ig_vw"] = "1536"
                self.s.cookies["ig_pr"] = "1.25"
                self.s.cookies["ig_vh"] = "772"
                self.s.cookies["ig_or"] = "landscape-primary"
                time.sleep(0.01)#5 * random.random())
                print(self.successfulLogin)

            if self.successfulLogin:
                self.r = self.s.get("https://www.instagram.com/")
                self.csrftoken = re.search('(?<="csrf_token":")\w+', self.r.text).group(0)
                self.s.cookies["csrftoken"] = self.csrftoken
                self.s.headers.update({"X-CSRFToken": self.csrftoken})
                self.finder = self.r.text.find(self.username)
                if self.finder != -1:
                    print("[{0}] Login succeeded!".format(self.username))
                    while True:
                            for x in self.targets:
                                # name = targets[x]
                                # checkUsername(name)
                                # x += 1
                                self.t = threading.Thread(target=self.checkUsername, args=(x,))
                                time.sleep(0.01)
                                self.t.start()
                    
                else:
                    print("Login error! Check your login data!")
            else:
                print("Login error! Connection error!")
                    
    def changeUsername(self, availableUsername):
       # try:
                        #self.t.stop()
                        self.username                    
                        self.url_edit = "https://www.instagram.com/accounts/edit/"
                        self.r = self.s.get("https://www.instagram.com/")
                        self.csrftoken = re.search('(?<="csrf_token":")\w+', self.r.text).group(0)
                        self.s.cookies["csrftoken"] = self.csrftoken
                        self.s.headers.update({"X-CSRFToken": self.csrftoken})
                        self.formData = {
                            'first_name': 'Vnes',
                            'email': self.email,
                            'username': availableUsername,
                            'phone_number': '',
                            'biography': '',
                            'external_url': '',
                            'chaining_enabled': 'on'
                        }
                        self.rf = self.s.post(self.url_edit, data=self.formData)
                        #self.t.start()
                        print(self.rf)
                        with open("config.json") as file:
                            update = json.load(file)
                            update['username'] = availableUsername
                            self.username = availableUsername
                            with open('config.json', 'w') as f:
                                json.dump(update, f)


       # except:
       #     pass


    def checkUsername(self, targetUsername):
        # global stop_threads
        # start monitoring the username
        # while not stop_threads:
        # while True:

        #try:
            #ip = Proxy.returned_proxies()
            usernameURL = 'https://www.instagram.com/web/search/topsearch/?' + urlencode({
                'context': 'blended', 'query': targetUsername, 'rank_token': '0.395359231827089', 'count': '1'
            })
            self.res = requests.get(usernameURL)#, proxies=ip, timeout=3)
            # time.sleep(1)
            self.json_data = json.loads(self.res.text)
            if self.res.status_code == 200:
                if len(self.json_data['users']) == 0:
                    print('[{0}] is available!!!'.format(targetUsername))
                    #try:
                    self.changeUsername(targetUsername)
                    #except:
                        #print('Error in Change in username. ')
                    time.sleep(2)
                    print('[{0}] Completed Turbo Killing Thread'.format(targetUsername))
                    #time.sleep(2)
                    with open('Found_Names.txt', 'a') as f:
                        f.write(targetUsername + '\n')
                    #exit()
                    # system.exit()
                else:
                    print('[{0}] is not available'.format(targetUsername))

        #except:
        #    pass


if __name__ == '__main__':
    checker = Checker()
    print("Instagram Username Taker")
    print("=" * 60)
    print("Username: {}".format(checker.username))
    print("Password: {}".format("*" * len(checker.password)))
    print("# Of Targets: {}".format(str(len(checker.targets))))
    # print(f"Endpoint: {endpoint}")
    print("=" * 60)
    print("")
    tin = input("Would you like to start (y/n)?: ")
    if tin.lower() == "y":
        checker.login()
    else:
        exit()

    # print(targets)
    #stop_threads = False
    
```
